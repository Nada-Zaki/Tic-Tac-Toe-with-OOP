```python
class TicTacToe:
    def __init__(self):
        print('Game created')

         
    # Function to print out the board    
    def display_board(self,board):
        self.board = board
        self.i=0
        while self.i<7:
            print('|',self.board[self.i],'|', self.board[self.i+1],'|', self.board[self.i+2], '|')
            self.i+=3
            
            
    # Function to take in a player input and assign their marker as 'X' or 'O'        
    def player_input(self):
        self.marker = input("Please pick a marker 'X' or 'O' ")
        self.marker = self.marker.upper()
        while(self.marker!='X' and self.marker!='O'):
            self.marker = input("please pick a correct marker ")
            self.marker = self.marker.upper()
        return self.marker
    
    
    # Function to take in the board list object, a marker ('X' or 'O'), 
    # and a desired position (number 1-9) and assign it to the board
    def place_marker(self,board, marker, position):
        self.board = board
        self.marker = marker
        self.position = position
        self.board[self.position-1] = self.marker
        
        
    # Function to take in a board and a mark (X or O) and then check to see if that mark has won   
    def win_check(self,board, mark):
        self.board = board
        self.mark = mark
        if(self.board[0]==self.mark and self.board[1]==self.mark and self.board[2]==self.mark):
            return True
        elif(self.board[3]==self.mark and self.board[4]==self.mark and self.board[5]==self.mark):
            return True
        elif(self.board[6]==self.mark and self.board[7]==self.mark and self.board[8]==self.mark):
            return True
        elif(self.board[0]==self.mark and self.board[3]==self.mark and self.board[6]==self.mark):
            return True
        elif(self.board[1]==self.mark and self.board[4]==self.mark and self.board[7]==self.mark):
            return True
        elif(self.board[2]==self.mark and self.board[5]==self.mark and self.board[8]==self.mark):
            return True
        elif(self.board[0]==self.mark and self.board[4]==self.mark and self.board[8]==self.mark):
            return True
        elif(self.board[2]==self.mark and self.board[4]==self.mark and self.board[6]==self.mark):
            return True
        else:
            return False
        
    
    # Function to use the random module to randomly decide which player goes first
    def choose_first(self):
        import random
        self.first_player = random.randint(0,1)
        if(self.first_player==0):
            return('player 1')
        else:
            return('player 2')
        
        
    # Function to return a boolean indicating whether a space on the board is freely available
    def space_check(self,board, position):
        self.board = board
        self.position = position
        if(self.board[self.position-1]=='*'):
            return True
        else:
            return False
        
        
    # Function to check if the board is full and return a boolean value. True if full, False otherwise   
    def full_board_check(self,board):
        self.board = board
        self.x=0
        for i in range(0,8):
            if(self.board[i]=='*'):
                self.x=1
                break
        if(self.x==1):
            return False
        else:
            return True
        
        
    # Function to ask for a player's next position (as a number 1-9) and then check 
    # if it's a free position. If it is, then return the position for later use   
    def player_choice(self,board):
        self.board = board
        while True:
            try:
                self.position = int(input('Your turn, enter number from 1 to 9 '))
            except:
                print('An error occurred! please try again!')
                continue
            else:
                break
        while not self.space_check(self.board,self.position):
            while True:
                try:
                    self.position = int(input('Please enter another number '))
                except:
                    print('An error occurred! please try again!')
                    continue
                else:
                    break
        return self.position
    
    
    # Function to ask the player if they want to play again 
    # and return a boolean True if they do want to play again
    def replay(self):
        self.answer = input("Do you want to play again? type 'yes' or 'no'")
        if(self.answer=='yes'):
            return True
        else:
            return False
```

```python
while True:
    # Set the game up here
    my_game = TicTacToe()
    print('Welcome to Tic Tac Toe!')
    board = ['*','*','*','*','*','*','*','*','*']
    first_player = my_game.choose_first()
    if(first_player == 'player 1'):
        print('player 1 goes first')
        player1_marker = my_game.player_input()
        if(player1_marker=='X'):
            player2_marker='O'
        elif(player1_marker=='O'):
            player2_marker='X'
        
        
    else:
        print('player 2 goes first')
        player2_marker = my_game.player_input()
        if(player2_marker=='X'):
            player1_marker='O'
        elif(player2_marker=='O'):
            player1_marker='X'
       
        
    while not my_game.full_board_check(board):
        if(first_player == 'player 1'):
            #Player 1 Turn
            player1_position = my_game.player_choice(board)
            my_game.place_marker(board, player1_marker, player1_position)
            my_game.display_board(board)
            if(my_game.win_check(board, player1_marker)):
                print('player 1 won')
                break
        
            # Player2's turn.
            player2_position = my_game.player_choice(board)
            my_game.place_marker(board, player2_marker, player2_position)
            my_game.display_board(board)
            if(my_game.win_check(board, player2_marker)):
                print('player 2 won')
                break
        else:
            # Player2's turn.
            player2_position = my_game.player_choice(board)
            my_game.place_marker(board, player2_marker, player2_position)
            my_game.display_board(board)
            if(my_game.win_check(board, player2_marker)):
                print('player 2 won')
                break
            
            #Player 1 Turn
            player1_position = my_game.player_choice(board)
            my_game.place_marker(board, player1_marker, player1_position)
            my_game.display_board(board)
            if(my_game.win_check(board, player1_marker)):
                print('player 1 won')
                break
                
    if not my_game.replay():
        break       
```
